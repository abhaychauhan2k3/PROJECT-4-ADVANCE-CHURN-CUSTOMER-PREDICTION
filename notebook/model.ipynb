Finding Binary, Nominal, and Ordinal cols

```python
df.head()
```

```python
object_cols2 = list(df.select_dtypes(include="object").columns)
df[object_cols2].nunique()
```

I have-
6 Binary cols 
11 nominal or ordinal cols 

now for binary cols we can do encoding -> 0/1 encoding

```python
binary_cols = df.select_dtypes(include = "object").columns[df.select_dtypes(include = "object").nunique() == 2].tolist()
binary_cols
```

```python
mapping_dict = {'Yes' : 1 , 'No': 0}
for column in binary_cols:
    if(column != "gender"):
        df[column] = df[column].map(mapping_dict) 
df.head()

```

```python
mapping_dict2 = {'Male' : 1 , 'Female': 0}
df['gender'] = df['gender'].map(mapping_dict2)
df.head()
```

```python
df = df.drop(columns = ["customerID"])
df.head()
```

```python
Nominal_cols = list(df.select_dtypes(include = "object").columns)
Nominal_cols
```

```python
for col in Nominal_cols:
    print(df[col].value_counts())
```

Here, Contract column is an Ordinal Col and rest object columns are Nominal cols
So, for Nominal Cols - using one hot encoding..

```python
Nominal_cols.remove("Contract")
Nominal_cols
```

Applying One hot encoding for all the nominal cols : to avoide artificial ordinal relationships

```python
df2 = pd.get_dummies(df, columns = Nominal_cols)
df2
```

Encoding ordinal column: Contract -> maintaing their inherent order

```python
ordinal_dict_mapping = {"Month-to-month": 0, "One year": 1, "Two year": 2}
df2['Contract'] = df['Contract'].map(ordinal_dict_mapping)
df2.head()
```

```python
bool_cols = list(df2.select_dtypes(include="bool").columns)
bool_cols
```

```python
for col in bool_cols:
    df2[col] = df2[col].astype(int)
```

```python
df2.head()
```

```python
df2.select_dtypes(include="bool").columns
```

Hence, dataset is Encoded Completely!!!
