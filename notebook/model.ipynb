Fixing Data Types of the Numeric columns stored as String or Objects


```python
#looking data types of each columns 
df.info()
```

```python
#put all those object types cols in list named object_cols
object_cols = list(df.select_dtypes(include="object").columns)
```

```python
#printing all those 18 cols and find out which col contains numeric values in string format 
df[object_cols].head()
```

Only TotalCharges is a column which contains numeric value but is of the object type, so Fixing its dataType

```python
df['TotalCharges'].isnull().sum()
```

Converting TotalCharges col to numeric type 

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors = "coerce")
```

```python
df['TotalCharges'].info()
```

```python
#checking for NaN in this col
df['TotalCharges'].isnull().sum()
```

```python
#fill these nan values with 0 
df['TotalCharges'].fillna(0, inplace = True)
```
